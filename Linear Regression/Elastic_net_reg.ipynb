```python
# importing

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression
```

```python
# Generate random data
X, y = make_regression(n_features=1, random_state=0, noise=10)
```

```python
# splitting

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train.shape
```

```python
# model

# Create the Elastic Net model
alpha = 0.5
enet = ElasticNet(random_state=0, )

# Fit the model to the training data
enet.fit(X, y)
```

```python
# predicting
y_pred = enet.predict(X)
```

```python
# evaluating

mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

X.shape, y.shape, y_pred.shape
```

```python
# visualising

# plt.scatter(X, y, color='blue', label='Actual')
plt.scatter(X[:,0], y, color='red', label='Regression Line')
plt.plot(X[:,0], y_pred, color='blue', label='Regression Line')
plt.legend()
plt.title('Elastic Net Regression')
plt.show()
```
