### Lasso Regression

```python
# importing 

from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# data

X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=0)
```

```python
# Splitting

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# model

model_1 = Lasso(alpha=1)
model_7 = Lasso(alpha=7)
model_20 = Lasso(alpha=20)

# fit
model_1.fit(X, y)
model_7.fit(X, y)
model_20.fit(X, y)
```

```python
# prediction

y_pred1 = model_1.predict(X_train)
y_pred7 = model_7.predict(X_train)
y_pred20 = model_20.predict(X_train)
```

```python
# Plot the data and the Lasso regression line

plt.scatter(X_train, y_train, label='Data')
plt.plot(X_train, y_pred1, color='red', linewidth=3, label='Lasso_1 Regression')
plt.plot(X_train, y_pred7, color='green', linewidth=3, label='Lasso_7 Regression')
plt.plot(X_train, y_pred20, color='black', linewidth=3, label='Lasso_20 Regression')

plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
