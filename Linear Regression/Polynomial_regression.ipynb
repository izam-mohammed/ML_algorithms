```python
# importing

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
# Generate example data
np.random.seed(0)
X, y = make_regression(n_samples=100, n_features=1, noise=30)

# Reshape the data (required for single feature)
X = X.reshape(-1, 1)
```

```python
# Create polynomial features
degree = 3  # Change the degree of the polynomial as needed
poly_features = PolynomialFeatures(degree=5)
X_poly = poly_features.fit_transform(X)

# Create and fit the polynomial regression model
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
X_poly.shape
```

```python
# predicting

y_pred = poly_reg.predict(X_poly)
```

```python
# Sort the data for smoother visualization
X_sorted = np.sort(X, axis=0)
y_pred_sorted = poly_reg.predict(poly_features.transform(X_sorted))

plt.scatter(X, y, label='Data', color='blue')
plt.plot(X_sorted, y_pred_sorted, label='Polynomial Regression', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Polynomial Regression')

plt.show()
```
