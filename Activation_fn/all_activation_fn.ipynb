```python
# importing 

import numpy as np
import matplotlib.pyplot as plt
```

```python
# data

X = np.linspace(-6, 6, 100)
```

```python
# plotting

plt.plot(X, X, color = "green")
```

```python
# step function

def step(x):
    return np.array([1 if i>0 else 0 for i in x])
```

```python
#plotting

vals = step(X)
plt.plot(X, vals)
```

```python
# sigmoid

def sigmoid(x):
    return 1/ (1+np.exp(-x))
```

```python
# plot

vals = sigmoid(X)
plt.plot(X, vals)
```

```python
# tanh

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
vals = tanh(X)
plt.plot(X, vals)
```

```python
# relu

def relu(x):
    return np.maximum(0, x)
```

```python
# plotting

vals = relu(X)
plt.plot(X, vals)
```

```python
# leaky relu

def leaky_relu(x):
    return np.maximum(x*0.1, x)
```

```python
# plotting

vals = leaky_relu(X)
plt.plot(X, vals)
```

```python
# softmax

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
# plot

vals = softmax(X)
plt.plot(X, vals)
```
